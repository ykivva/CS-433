```python
def print_nans_per_column(data):
    counts = np.isnan(tX).sum(axis=0)
    fraction = counts / data.shape[0]
    print('%10s%10s%10s' % ('col', 'nans', 'fraction'))
    for i in range(tX.shape[1]):
        print('%10d%10d%10f' % (i, counts[i], fraction[i]))


def plot_feature_hists(data, bins=200):
    fig, ax = plt.subplots(7,5, figsize=(18,20))
    for row in range(7):
        for col in range(5):
            n = row*5+col
            if n < data.shape[1]:
                ax[row,col].hist(data[:,n], bins=bins)
                ax[row,col].set_title(str(n))
    plt.show()
```
